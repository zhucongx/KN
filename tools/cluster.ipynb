```python
import json
import numpy
import matplotlib.pyplot as plt
```

```python
with open('clusters_info.json') as f:
    data = json.load(f)
```

```python
cluster_num = []
time = []
atoms = []
size = []
ra = []
for i in data:
    time.append(i["time"]/10)
    cluster_num.append(len(i["clusters"]))
    ct=0
    for j in i["clusters"]:
        ct+=j[0]
        ct+=j[1]
        ct+=j[2]
    atoms.append(ct)
    size.append(ct/len(i["clusters"]))
```

```python
fig, ax = plt.subplots()
ax.plot(time[1:-1:10], cluster_num[1:-1:10])

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(20)
ax.set_xlabel('simulated time (sec)', fontdict={'family' : 'Times New Roman', 'size':25})

ax.set_ylabel('num of cluster',fontdict={'family' : 'Times New Roman', 'size':25})
```

```python
fig, ax = plt.subplots()
ax.plot(time[1:-1:10], atoms[1:-1:10])

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(20)

ax.set_xlabel('simulated time (sec)', fontdict={'family' : 'Times New Roman', 'size':25})
ax.set_ylabel('num of atoms in cluster', fontdict={'family' : 'Times New Roman', 'size':25})
```

```python
fig, ax = plt.subplots()
ax.plot(time[1:-1:10], size[1:-1:10])

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(20) 
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(20)

ax.set_xlabel('simulated time (sec)', fontdict={'family' : 'Times New Roman', 'size':25})
ax.set_ylabel('average cluster size', fontdict={'family' : 'Times New Roman', 'size':25})
```
